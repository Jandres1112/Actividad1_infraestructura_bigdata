Lectura de los Datos con pandas:
Se hace un primer analisis y se convierte a excel  

```python
import pandas as pd

# Assuming the CSV file is located at '/content/Centros_Digitales_Rurales_20260322.csv'
# If the file name or path is different, please update it accordingly.
df_datos = pd.read_excel("/content/Centros_Digitales_Rurales_20260322.xlsx")


# Display the first 5 rows of the DataFrame
display(df_datos.head())
```

Esploración de datos

```python
#cantidad de registros duplicados

duplicados_multi = df_datos[df_datos.duplicated(subset=['FECHA_CORTE', 'DEPARTAMENTO', 'MUNICIPIO', 'PRIORIZACION', 'ZONA',
       'DIFICULTADACCESO', 'NOMBRE_CENTRO_POBLADO', 'TIPO_SITIO',
       'TIPO_CONECTIVIDAD', 'NOMBRE_INSTITUCION_EDUCATIVA',
       'NOMBRE_SEDE_EDUCATIVA', 'ESTADOS', 'TIPO_ENERGIA', 'DETALLE_SITIO',
       'USUARIOS_ACTIVOS_MES', 'VELOCIDAD_CONEXION_SUBIDA',
       'VELOCIDAD_CONEXION_BAJADA', 'TRAFICO_MENSUAL_SUBIDA',
       'TRAFICO_MENSUAL_BAJADA', 'INVERSION', 'META'])]
print(duplicados_multi)
```

```python
# Conteo de nulos por columna
print("el total de datos nulos por columna es:", df_datos.isnull().sum())
#Dimenciones del df
print("Dimensiones del DataFrame filas x columnas:", df_datos.shape)
```

```python
#total de datos nulos en el dataframe
total_nulos = df_datos.isnull().sum().sum()
print("Total de valores nulos en el DataFrame:", total_nulos)

```

```python
# 3. Mostrar filas con al menos un valor nulo
print("filas con al menos un dato nulo:", df_datos[df_datos.isnull().any(axis=1)])
```

```python
df_limpio = df_datos.copy()
df_limpio['NOMBRE_INSTITUCION_EDUCATIVA'] = df_limpio['NOMBRE_INSTITUCION_EDUCATIVA'].fillna("Desconocido")
df_limpio['TIPO_ENERGIA'] = df_limpio['TIPO_ENERGIA'].fillna("Desconocido")
```

```python
#Validacion Datos nulos

# 1. Ver si existen nulos en todo el DataFrame
print("Valores nulos en el dataframe:", df_limpio.isnull().any().any())
# 2. Contar nulos por columna
print("cantidad de valores nulos en el dataframe:", df_limpio.isnull().sum())
#copiar el archivo
df_limpio.to_excel("df_limpio.xlsx", index=False)
```

```python
#Validacion tipo de dato
print("las columnas del dataframe limpio:", df_limpio.info())
```

```python
import pandas as pd
#Generar archivo auditoria
log = []

#----Cantidad datos antes de limpieza----
registros_iniciales = len(df_datos)
log.append(f"Registros iniciales: {registros_iniciales}")

#----Cantidad de datos duplicados----
log.append(f"cantidad de registros duplicados: {duplicados_multi}")

#----cantidad de datos nulos----
log.append(f"Total de valores nulos en el DataFrame: {total_nulos}")


# --- Celda final: generar archivo de auditoría ---
with open("limpio_report.txt", "w", encoding="utf-8") as f:
    f.write("=== Informe de Limpieza de Datos ===\n\n")
    for linea in log:
        f.write(linea + "\n")



```
