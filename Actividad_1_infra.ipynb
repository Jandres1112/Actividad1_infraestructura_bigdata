```python
import requests
import sqlite3
import pandas as pd
import openpyxl

#Extracción de datos
API_URL = "https://jsonplaceholder.typicode.com/posts"


def obtener_datos(url):
    response = requests.get(url)
    if response.status_code == 200:
        return  response.json()
    else:
        raise Exception(f"Error al conectar con la API: {response.status_code}")

#guardar en la base

def guardar_en_base(datos):
    conn = sqlite3.connect('datos_extraidos.db')
    cursor = conn.cursor()

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS POST (
            id INTEGER PRIMARY KEY,
            id_usuario INTEGER,
            titulo TEXT,
            comentario TEXT
        )
    ''')

#insertar datos
    for item in datos:
        cursor.execute('''
            INSERT OR REPLACE INTO POST
            (id, id_usuario, titulo, comentario)
            VALUES (?, ?, ?, ?)
        ''', (
            item['id'],
            item['userId'],
            item['title'],
            item['body']

        ))

    conn.commit()
    conn.close()
# 3. GENERAR EVIDENCIAS (Pandas y Auditoría)
def generar_evidencias(datos):
    # --- Archivo Excel/CSV con Pandas ---
    conn = sqlite3.connect('datos_extraidos.db')
    df_base = pd.read_sql_query("SELECT * FROM POST", conn)
    df_base.to_csv('comentarios.csv', index=False)
    conn.close()

    # --- Archivo de Auditoría .txt ---
    registros_en_api = len(datos)
    registros_en_db = len(df_base)

    with open('logs.txt', 'w', encoding='utf-8') as f:
        f.write("RESUMEN DE AUDITORÍA\n")
        f.write("====================\n")
        f.write(f"Registros extraídos de la API: {registros_en_api}\n")
        f.write(f"Registros guardados en la DB: {registros_en_db}\n")

        if registros_en_api == registros_en_db:
            f.write("\nESTADO: Sincronización Exitosa.\n")
        else:
            f.write("\nESTADO: Discrepancia detectada.\n")

# EJECUCIÓN DEL PROCESO
try:
    print("Iniciando proceso...")
    datos = obtener_datos(API_URL)

    print("Guardando en base de datos...")
    guardar_en_base(datos)

    print("Generando evidencias (CSV y Auditoría)...")
    generar_evidencias(datos)

    print("¡Proceso completado con éxito!")
except Exception as e:
    print(f"Ocurrió un error: {e}")
```
